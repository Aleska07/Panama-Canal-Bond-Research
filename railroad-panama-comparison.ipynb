```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load railroad data
railroad_data = pd.read_csv("/Users/aleskamedrano/Desktop/GitHub/HistoricalEconomics/code/personal/aleska/Project/Panama Canal/Railroad_data.csv")
railroad_month = pd.to_datetime(railroad_data["DATE"])
railroad_price = railroad_data["Price"].values.astype(float) 

# Load Panama Canal bond data (series 1906)
panama_bond_data = pd.read_csv("/Users/aleskamedrano/Desktop/GitHub/HistoricalEconomics/code/personal/aleska/Project/transposed_file.csv")
panama_bond_price_1906 = panama_bond_data["20132"].values.astype(float) 
panama_bond_price_1908 = panama_bond_data["20133"].values.astype(float) 
panama_bond_price_1911 = panama_bond_data["20134"].values.astype(float) 
panama_bond_date = pd.to_datetime(panama_bond_data["Date"].values)

# Create subplots with shared y-axis for bond prices
fig, axs = plt.subplots(1, 2, figsize=(14,6), sharey=True)

# Plot railroad data
axs[0].plot(railroad_month, railroad_price, color='black', label='Railroad Price')

# Events to mark on the railroad plot
events = {
    "Fluctuations Change": "1917-04-01",
}

# Plot markers and annotations for the events
for event, date in events.items():
    event_date = pd.to_datetime(date)
    if event_date in railroad_month.values:
        idx = railroad_month[railroad_month == event_date].index[0]
        axs[0].scatter(event_date, railroad_price[idx], color='red')  # Add marker
        axs[0].annotate(event, (event_date, railroad_price[idx]), 
                        textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

axs[0].set_title("Railroad Bonds (1913-1923)")
axs[0].set_xlabel("Year")
axs[0].set_ylabel("Price")
axs[0].set_xlim(pd.to_datetime('1913-01-01'), pd.to_datetime('1923-01-01')) 

# Plot bond data for Panama Canal
axs[1].plot(panama_bond_date, panama_bond_price_1906, color='black', label='Series 1906')
axs[1].plot(panama_bond_date, panama_bond_price_1908, color='red', label='Series 1908')
axs[1].plot(panama_bond_date, panama_bond_price_1911, color='blue', label='Series 1911')

# Plot markers and annotations for the same events
for event, date in events.items():
    event_date = pd.to_datetime(date)
    if event_date in panama_bond_date.values:
        idx = panama_bond_date[panama_bond_date == event_date].index[0]
        axs[1].scatter(event_date, panama_bond_price_1906[idx], color='red')  # Example marker
        axs[1].annotate(event, (event_date, panama_bond_price_1906[idx]), 
                        textcoords="offset points", xytext=(0,10), ha='center', fontsize=8)

axs[1].set_title("Panama Canal Bonds (1913-1923)")
axs[1].set_xlabel("Year")
axs[1].set_ylabel("Price")
axs[1].set_xlim(pd.to_datetime('1913-01-01'), pd.to_datetime('1923-01-01')) 

# Adjust layout and add a separate legend outside the plots
fig.tight_layout(rect=[0, 0, 0.85, 1])  # Make room on the right for the legend

# Function to Create a combined legend outside the plots
fig.legend(loc='center left', bbox_to_anchor=(0.9, 0.5), fontsize='small')

fig.savefig('Railroad_Panama.png', bbox_inches='tight', dpi=300)

# Display the plot
plt.show()
```
