```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize

class Coupon_Bond:
    def get_price(self, coupon, face_value, int_rate, years, freq=4):
        total_coupons_pv = self.get_coupons_pv(coupon, int_rate, years, freq)
        face_value_pv = self.get_face_value_pv(face_value, int_rate, years)
        result = total_coupons_pv + face_value_pv
        return result

    @staticmethod
    def get_face_value_pv(face_value, int_rate, years):
        fvpv = face_value / (1 + int_rate)**years
        return fvpv

    def get_coupons_pv(self, coupon, int_rate, years, freq=4):
        pv = 0
        for period in range(years * freq):
            pv += self.get_coupon_pv(coupon, int_rate, period + 1, freq)
        return pv

    @staticmethod
    def get_coupon_pv(coupon, int_rate, period, freq):
        pv = coupon / (1 + int_rate / freq)**period
        return pv

    def get_ytm(self, bond_price, face_value, coupon, years, freq=4, estimate=0.05):
        get_yield = lambda int_rate: self.get_price(coupon, face_value, int_rate, years, freq) - bond_price
        return scipy.optimize.newton(get_yield, estimate)

    def plot_ytm_vs_price(self, face_value, coupon, years, freq=4, estimate=0.05):
        # Define range of bond prices
        bond_prices = np.linspace(100, 110, 40)  # Adjust range as needed
        
        # Calculate YTM for each bond price
        ytms = []
        for price in bond_prices:
            ytm = self.get_ytm(price, face_value, coupon, years, freq, estimate)
            ytms.append(ytm)
        
        # Plotting
        plt.figure(figsize=(10, 6))
        plt.plot(bond_prices, ytms, marker='o', linestyle='-', color='b', label='Yield to Maturity')
        plt.title('Yield to Maturity vs. Bond Price')
        plt.xlabel('Bond Price')
        plt.ylabel('Yield to Maturity')
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

# Example usage
coupon_bond_calculator = Coupon_Bond()
face_value = 1000
coupon = 0.02
years = 30
freq = 4
estimate = 0.05

coupon_bond_calculator.plot_ytm_vs_price(face_value, coupon, years, freq, estimate)
```
